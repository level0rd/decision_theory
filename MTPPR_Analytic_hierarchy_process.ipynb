## Школьный совет

```python
import pandas as pd
from scipy. stats import gmean

a_c=[['B',1,1],
   ['P',1,1]]

header = ['','B', 'P']
df_c = pd.DataFrame(a_c,columns=header)  

df_c['yi'] = gmean([df_c['B'], df_c['P']]) # среднее геометрическое
sum_yi = df_c['yi'].sum()
df_c['yin'] = df_c['yi']/sum_yi

#sum_yin = df_c['yin'].sum()
# df['λi'] = 0
# df['λi'][0] = df['B'].sum() * df['yin'][0]
# df['λi'][1] = df['P'].sum() * df['yin'][1]

# λmax =  df['λi'].sum()
# n = count_row = df.shape[0]
# ic = (λmax - n)/(n - 1)
# cc = 0.00001
# oc = ic/cc

# print(f'λmax = {λmax}')
# print(f'IC = {ic}')
# print(f'CC = {cc}')
# print(f'OC = {oc}')

df_cprt=df_c.style.hide(axis='index')
df_cprt
```

### Безопасность

```python
a_cb=[['F',1,2],
   ['M',1/2,1]]

header = ['','F', 'M']
df_cb = pd.DataFrame(a_cb,columns=header)  

df_cb['yi'] = gmean([df_cb['F'], df_cb['M']])
sum_yi = df_cb['yi'].sum()
df_cb['yin'] = df_cb['yi']/sum_yi

df_cbprt=df_cb.style.hide(axis='index')
df_cbprt
```

## Расчет глобальных приоритетов

### Школьный совет

```python
cres = [['',df_c['yin'][0], df_c['yin'][1]],
       ['F',df_cb['yin'][0],df_cp['yin'][0]],
       ['M',df_cb['yin'][1],df_cp['yin'][1]]]

header = ['','B', 'P']
df_cres = pd.DataFrame(cres,columns=header) 

df_cres['Ai'] = ''
df_cres['Ai'][1] = df_c['yin'][0] * df_cb['yin'][0] + df_c['yin'][1] * df_cp['yin'][0]
df_cres['Ai'][2] = df_cb['yin'][1] * df_c['yin'][0] + df_cp['yin'][1] * df_c['yin'][1]
```

```python
df_cres=df_cres.style.hide(axis='index')
df_cres
```
