# Узнаем как часто встречается каждая мера защиты при использовании модели нарушителя с базовыми возможностями

```python
import pandas as pd

df1 = pd.read_excel('df1.xlsx')
df1
```

```python
df1_org = pd.read_excel('Перечень сформированных угроз 1.xlsx')
df1_org
```

```python
# Подсчитываем количество вхождений каждого значения в df2
counts = df1_org['Меры защиты'].value_counts()

# Используем map для создания нового столбца в df1, основанного на counts
df1['Число вхождений'] = df1_org['Меры защиты'].map(counts)

df1
```

```python
# Находим уникальные значения в df2, которые также есть в df1
unique_values = df1_org[df1_org['Меры защиты'].isin(df1['Меры защиты'])]['Меры защиты'].unique()

# Для каждого уникального значения подсчитываем количество его вхождений в df2
counts = {value: (df1_org['Меры защиты'] == value).sum() for value in unique_values}

# Используем map для создания нового столбца в df1, основанного на counts
df1['Число вхождений'] = df1['Меры защиты'].map(counts)

df1
```

```python
df1_sorted = df1.sort_values('Число вхождений', ascending=False)
df1_sorted
```

```python
# Проверка
print(df1['Число вхождений'].sum())
print(len(df1_org))
```

```python
# Оставляем в df только два последних столбца
df1_sorted = df1_sorted.iloc[:, -2:]
df1_sorted
```

```python
#df1_sorted.iloc[0, 0]
substrings = df1_sorted.iloc[0, 0].split(' ;_x000d_\n')
substrings
```

```python

# Создаем первый DataFrame
res_df = pd.DataFrame({
    'Меры защиты': None,
    'Число вхождений': [0]
})

for j in range(len(df1_sorted)):

    substrings = df1_sorted.iloc[j, 0].split(';_x000d_')

    for i in range(len(substrings)):
        new_row = pd.DataFrame({
            'Меры защиты': [substrings[i]],
            'Число вхождений': [df1_sorted.iloc[j, 1]]
        })

        res_df = pd.concat([res_df, new_row], ignore_index=True)

res_df
```

```python
# Удаляем первую строку
res_df = res_df.drop(res_df.index[0])
```

```python
# Группируем df по 'Первый столбец' и суммируем значения в 'Второй столбец'
result = res_df.groupby('Меры защиты', as_index=False)['Число вхождений'].sum()
result
```

```python
result_sorted = result.sort_values('Число вхождений', ascending=False)
result_sorted
```

```python
# Удаляем пробел в начале строки
result_sorted['Меры защиты'] = result_sorted['Меры защиты'].apply(lambda s: s.lstrip())
result_sorted
```

```python
duplicates = result_sorted.duplicated(subset='Меры защиты')

# Если вы хотите увидеть дубликаты, вы можете сделать следующее:
result_duplicates = result_sorted[duplicates]
result_duplicates
```

```python
# Группируем df по 'Первый столбец' и суммируем значения в 'Второй столбец'
result_sorted = result_sorted.groupby('Меры защиты', as_index=False)['Число вхождений'].sum()
result_sorted
```

```python
result_sorted = result_sorted.sort_values('Число вхождений', ascending=False)
result_sorted
```

```python
result_sorted['Меры защиты'] = result_sorted['Меры защиты'].str.extract('([А-ЯЁ]{2,}\.\d+\.\d+)')  # Оставить только абревиатуры
result_sorted
```
